## Get bounding boxes from parking lot images using YOLOv8n

```python
# Load images from subdirs, including date subfolders
import os

image_data = []

project_dir = os.path.abspath(os.getcwd())
image_dir = os.path.join(project_dir, 'PKLotYoloData')

for xml_type in os.listdir(image_dir):
    xml_type_path = os.path.join(image_dir, xml_type)
    print(f"XML type path: {xml_type_path}   ")
    for parking_lot in os.listdir(xml_type_path):
        parking_lot_path = os.path.join(xml_type_path, parking_lot)
        print(f"Parking lot path: {parking_lot_path}   ")
        if os.path.isdir(parking_lot_path):  # Ensure it's a directory
            for weather_condition in os.listdir(parking_lot_path):
                weather_path = os.path.join(parking_lot_path, weather_condition)
                print(f"\tWeather condition path: {weather_path}   ")
                if os.path.isdir(weather_path):  # Ensure it's a directory
                    for date_folder in os.listdir(weather_path):
                        date_path = os.path.join(weather_path, date_folder)
                        if os.path.isdir(date_path):  # Ensure it's a directory
                            for file in os.listdir(date_path):
                                if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                                    image_data.append({
                                        'parking_lot': parking_lot,
                                        'weather': weather_condition,
                                        'date': date_folder,
                                        'file_path': os.path.join(date_path, file)
                                    })

# Example: Print the first few entries
for entry in image_data[:5]:
    print(entry)
```

```python
# Convert image_data to a pandas DataFrame
import pandas as pd

# Create a DataFrame from the image_data list
df = pd.DataFrame(image_data)

# Display the first few rows of the DataFrame
display(df)
```

```python
#Save dataframe to CSV
df.to_csv('file_names_dataframe.csv', index=False)
```
